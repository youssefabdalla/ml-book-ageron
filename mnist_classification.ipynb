# classifing digits in MNIST dataset

## loading dataset

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist.data, mnist.target
```

```python
# function to plot an image
import numpy as np
import matplotlib.pyplot as plt
def plot_digit(image_data:np.ndarray):
    plt.axis('off')
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap='binary')
```

```python
some_digit = X[0]
plot_digit(some_digit)
plt.show()
y[0]
```

## splitting into training and validation data sets

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
def plot_100_digits(X):
    num_rows = 10
    num_cols = 10
    _, axes = plt.subplots(num_rows, num_cols, figsize=(10, 10))  # Adjust figsize as needed

    for i in range(num_rows * num_cols):
        row = i // num_cols
        col = i % num_cols
        image_data = X[i] # Get the i-th image data
        image = image_data.reshape(28, 28)
        axes[row,   ].imshow(image, cmap='binary')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
plot_100_digits(X_train)
```

## The 5 classifier

```python
#Stochatic gradient descent
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```
